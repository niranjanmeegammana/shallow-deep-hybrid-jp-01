```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Data
data = {
    "Model": ["Concat40", "Maximum40", "WgtAv40", "Deep40", "Shallow20"],
    "Accuracy (%)": [98.0, 98.0, 98.0, 97.0, 93.0],
    "FPR": [1, 1, 2, 3, 3],  # Low=1, Moderate=2, High=3
    "FNR": [1, 2, 1, 1, 3],  # Low=1, Moderate=2, High=3
    "Robustness": [4, 3, 3, 2, 1]  # Very High=4, High=3, Moderate=2, Weak=1
}

# Convert to DataFrame
df = pd.DataFrame(data)
df.set_index("Model", inplace=True)

# Normalize data to range [0, 1]
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns, index=df.index)

# Create heatmap with normalized data
plt.figure(figsize=(8, 6))
sns.heatmap(df_normalized, annot=df, cmap="Blues", linewidths=0.5, fmt=".1f")

# Title and labels
#plt.title("Normalized Performance Trade-offs for Adversarial Robustness", fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.savefig("performance_heat_map.png", dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ["Concat40", "WgtAv40", "Maximum40", "Minimum40", "Subtract40", "Multiply40"]
categories = ["Accuracy Rank", "Precision Rank", "Adversarial Robustness Rank"]
values = [
    [1, 1, 1],
    [2, 2, 2],
    [3, 3, 3],
    [4, 4, 4],
    [5, 5, 5],
    [6, 6, 6]
]

# Number of variables
num_vars = len(categories)

# Compute angles for radar chart
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Complete the loop for closed shape
angles += angles[:1]

# Create figure
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Set font
plt.rcParams.update({'font.family': 'serif', 'font.size': 14})

# Plot each model
for i, model in enumerate(models):
    data = values[i] + values[i][:1]  # Repeat first value for closing the shape
    ax.plot(angles, data, label=model, linewidth=2)
    ax.fill(angles, data, alpha=0.1)

# Set category labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

# Add title
#ax.set_title("Model Performance Comparison", fontsize=16)

# Add legend
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

# Save as high-resolution image
plt.savefig("radar_chart.png", dpi=300, bbox_inches='tight')

# Show plot
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Data for models
models = ["Shallow20", "Deep40", "WgtAv40", "Concat40", "Maximum40"]
cpu_usage = [12.8, 48.2, 56.7, 65.3, 68.9]  # CPU Usage in %
memory_usage = [75, 230, 275, 310, 320]  # Memory Usage in MB
adversarial_robustness = [1, 3, 4, 5, 4]  # Arbitrary robustness scale (1-5)

# Set up figure with 300 DPI
plt.figure(figsize=(7, 5), dpi=300)

# Create scatter plot
plt.scatter(cpu_usage, adversarial_robustness, color="b", s=100, edgecolors="black", label="CPU Usage")
plt.scatter(memory_usage, adversarial_robustness, color="r", s=100, edgecolors="black", label="Memory Usage")

# Annotate points with model names
for i, model in enumerate(models):
    plt.annotate(model, (cpu_usage[i] + 1, adversarial_robustness[i] - 0.1), fontsize=12, fontfamily="serif")

# Labels and title
plt.xlabel("Resource Usage (CPU % / Memory MB)", fontsize=14, fontfamily="serif")
plt.ylabel("Adversarial Robustness (1-5 Scale)", fontsize=14, fontfamily="serif")
plt.title("Figure 4.5: Resource Utilization vs. Adversarial Robustness", fontsize=14, fontfamily="serif")

# Grid, legend, and layout
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(fontsize=12, fontfamily="serif")
plt.yticks(fontsize=12, fontfamily="serif")
plt.legend(fontsize=12, frameon=False)

# Save figure with 300 DPI
plt.savefig("Figure_4.5_Resource_Utilization.png", dpi=300, bbox_inches="tight")

# Show the figure
plt.show()
```
